```python
from txtai.embeddings import Embeddings
import pandas as pd
```

```python
df = pd.read_csv("data/shakespeare.csv")
df
```

```python
lines = df.PlayerLine.tolist()
```

```python
embeddings = Embeddings({"path": "sentence-transformers/all-MiniLM-L6-v2"})
```

```python
embeddings.index([(i, line, None) for i, line in enumerate(lines)])
```

```python
embeddings.save("models/shakespeare")
```

```python
df.loc[69148]
```

```python
for index, score in embeddings.search("knife", 25):
    print(index)
    print(df.loc[index].Player)
    print(df.loc[index].PlayerLine)
    print(score)
    print()
```
